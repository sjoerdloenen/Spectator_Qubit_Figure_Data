```python
# import general packages
import os
import dill as pickle
pickle.settings['recurse'] = True

# Matplotlib
from matplotlib import pyplot as plt

# Import functions
%matplotlib inline
```

```python
axes_color =  '0.8'

plt.rcParams.update({'font.size': 9,
                     'font.family': 'sans-serif',
                     'font.sans-serif' : 'Helvetica',
                     'font.style':'normal',
                     'axes.grid' : False,
                     'axes.edgecolor': axes_color,
                     'ytick.color': '.15',
                     'xtick.color':'.15',
                     'lines.linewidth':1,
                    })

ytick_major_size = 3.5
plt.rcParams['ytick.major.size'] = ytick_major_size
plt.rcParams['ytick.major.width'] = 0.8

xtick_major_size = 3.5
plt.rcParams['xtick.major.size'] = xtick_major_size
plt.rcParams['xtick.major.width'] = 0.8
plt.rcParams['lines.markersize'] = 2.5

plt.rcParams['errorbar.capsize'] = 1.5
plt.rcParams['lines.markeredgewidth'] = 0.5


linewidth_default = plt.rcParams['lines.linewidth']
markersize_default = plt.rcParams['lines.markersize']
elinewidht_default = plt.rcParams['lines.markeredgewidth']
capsize_default = plt.rcParams['errorbar.capsize']
markeredgewidth_default = plt.rcParams['lines.markeredgewidth']
```

# Function to load the data

```python
def data_dict_from_pkl(DataName):
    FolderPath = os.getcwd()
    DataFilePath = os.path.join(FolderPath,DataName)
    
    DataFile = open(DataFilePath, 'rb')
    Data = pickle.load(DataFile, encoding = 'latin1')
    DataFile.close()

    return Data
```

# Load the data

```python
data_to_plot_xy = data_dict_from_pkl('data_to_plot_xy.pkl')
data_to_plot_z = data_dict_from_pkl('data_to_plot_z.pkl')
```

# Generate the plots XY from the datasets

```python
N_Data_Qbts = data_to_plot_xy[0]['N_Data_Qbts']
N_Spec_Configs = data_to_plot_xy[0]['N_Spec_Configs']


fig_multiplier = 1
fig_xy, ax_xy = plt.subplots(N_Spec_Configs, N_Data_Qbts, figsize = (18.3/2.54*fig_multiplier,N_Spec_Configs/2.*8.9/2.54*fig_multiplier))

N_msmts = len(data_to_plot_xy)
colorMap = ['black','grey']


for Cntr in range(N_msmts):

    N_Spec_Configs = data_to_plot_xy[Cntr]['N_Spec_Configs']
    N_Data_Qbts = data_to_plot_xy[Cntr]['N_Data_Qbts']
    RowNr = data_to_plot_xy[Cntr]['RowNr']
    ColNr = data_to_plot_xy[Cntr]['ColNr']
    title = data_to_plot_xy[Cntr]['title']

    if N_Data_Qbts == 1 or N_Spec_Configs == 1:
        Ax = ax_xy[Cntr]
    else:
        Ax = ax_xy[RowNr, ColNr]

    dataset_schemes = list(data_to_plot_xy[Cntr]['data'].keys())

    for schemeCntr, Scheme in enumerate(dataset_schemes):
        x = data_to_plot_xy[Cntr]['data'][Scheme]['x']
        y = data_to_plot_xy[Cntr]['data'][Scheme]['y']
        yerr = data_to_plot_xy[Cntr]['data'][Scheme]['yerr']
        x_fit = data_to_plot_xy[Cntr]['data'][Scheme]['x_fit']
        y_fit = data_to_plot_xy[Cntr]['data'][Scheme]['y_fit']
    
        Color = colorMap[schemeCntr]
        Ax.errorbar(x,y,yerr, fmt = 'o', color = Color, ecolor = Color, mfc = Color, mec = Color)
        Ax.plot(x_fit, y_fit, color = Color, label = Scheme) 


    # Setting properties of plot which are independent of the scheme implementation
    Ax.legend(fontsize = 7)
    Ax.set_ylim([0.45,1.05])
    Ax.grid(True)
    if RowNr == N_Spec_Configs-1: 
        xlabel = r'$\mathrm{N_{REA}}$'
        Ax.set_xlabel(xlabel)
    ylabel = '$\mathrm{F_{xy}}$' 
    Ax.set_ylabel(ylabel)
    Ax.title.set_text(title)
    Ax.title.set_fontsize(7)

    Ax.axhline(y=2./3, c = 'k', alpha = 0.5, linestyle = '--')

plt.tight_layout()

# FolderPath =r'...' 
# FigName = 'SI_Fig9.pdf'
# FigSavePath = os.path.join(FolderPath,FigName)

# plt.savefig(FigSavePath, bbox_inches='tight')# , transparent=True)
```
